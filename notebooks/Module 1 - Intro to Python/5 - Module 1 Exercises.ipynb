### 2.3
Write a function called `discretize_bmi` which takes a numeric BMI value to the following categories:
- **<18.5**
- **18.5-24.9**
- **25.0-29.9**
- **30-39.9**
- **>=40**

Then make a list called `bmis_discrete` which contains the categorized BMI values for `bmis`.

```python
def discretize_bmi(bmi):
    if bmi < 18.5:
        return "<18.5"
    elif 18.5 <= bmi <= 24.9:
        return "18.5-24.9"
    elif 25.0 <= bmi <= 29.9:
        return "25.0-29.9"
    elif 30.0 <= bmi <= 39.9:
        return "30-39.9"
    else:
        return ">=40"

bmis_discrete = [discretize_bmi(bmi) for bmi in bmis]
```
